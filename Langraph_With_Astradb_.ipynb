# Setup Query Router

```python
from typing import Literal
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.pydantic_v1 import BaseModel, Field

# Define router to direct queries to the appropriate data source
class RouteQuery(BaseModel):
    """Route a user query to the most relevant datasource."""

    datasource: Literal["vectorstore", "wiki_search", "stack_exchange", "arxiv"] = Field(
        ...,
        description="Route user questions to vectorstore, wikipedia, stack exchange, or arxiv search.",
    )
```

```python
from google.colab import userdata
from langchain_groq import ChatGroq # LLM with function call
import os

groq_api_key=userdata.get('groq_api_key')
os.environ["GROQ_API_KEY"]=groq_api_key
llm=ChatGroq(groq_api_key=groq_api_key,model_name="Gemma2-9b-It")

# Initialize prompt and set guidelines for routing
system = """You are an expert at routing user questions to the appropriate datasource.
The vectorstore contains documents related to agents, prompt engineering, and adversarial attacks.
Use the vectorstore for questions on these topics.
Use stack_exchange for technical programming questions or when seeking solutions to specific technical problems.
Use arxiv for questions about research papers, academic publications, or scientific literature.
Use wiki-search for general knowledge questions.

Guidelines:
- If the question mentions "code", "programming", "how to", "error", or specific technical terms, route to stack_exchange
- If the question includes terms like "research papers", "publications", "studies", "papers", or "academic", route to arxiv
- If the question is about agents, prompt engineering, or adversarial attacks, route to vectorstore
- For general knowledge questions, route to wiki_search"""

route_prompt = ChatPromptTemplate.from_messages([
    ("system", system),
    ("human", "{question}"),
])

# Define structured router and test with example queries
structured_llm_router = llm.with_structured_output(RouteQuery)
question_router = route_prompt | structured_llm_router

print(question_router.invoke({"question": "who is Sharukh Khan?"}))
print(question_router.invoke({"question": "What are the types of agent memory?"}))
print(question_router.invoke({"question": "How do I fix a NullPointerException in Java?"}))
print(question_router.invoke({"question": "What research papers exist about transformers in deep learning?"}))
```

# Graph and Workflow Setup

Define a data model for tracking the state of the graph-based workflow

```python
## Graph

from typing import List
from typing_extensions import TypedDict

class GraphState(TypedDict):
    """
    Represents the state of our graph.

    Attributes:
        question: question
        generation: LLM generation
        documents: list of documents
    """

    question: str
    generation: str
    documents: List[str]
```

Define retrieval functions for different data sources (VectorDB, Wikipedia, Stack Exchange, Arxiv)

Each function retrieves data based on the question and updates the GraphState

```python
from langchain.schema import Document

def retrieve(state):
    """
    Retrieve documents

    Args:
        state (dict): The current graph state

    Returns:
        state (dict): New key added to state, documents, that contains retrieved documents
    """
    print("---RETRIEVE---")
    question = state["question"]

    # Retrieval
    documents = retriever.invoke(question)
    return {"documents": documents, "question": question}
```

Wiki search function to fetch relevant articles

```python
def wiki_search(state):
    """
    wiki search based on the re-phrased question.

    Args:
        state (dict): The current graph state

    Returns:
        state (dict): Updates documents key with appended web results
    """

    print("---wikipedia---")
    print("---HELLO--")
    question = state["question"]
    print(question)

    # Wiki search
    docs = wiki.invoke({"query": question})
    #print(docs["summary"])
    wiki_results = docs
    wiki_results = Document(page_content=wiki_results)

    return {"documents": wiki_results, "question": question}
```

Stack Exchange search function to fetch technical answers

```python
# Stack Exchange search function
def stack_exchange_search(state):
    """
    Search Stack Exchange sites for technical questions and answers.

    Args:
        state (dict): The current graph state

    Returns:
        state (dict): Updates documents key with Stack Exchange results
    """
    print("---STACK EXCHANGE SEARCH---")
    question = state["question"]

    # Initialize Stack API (using Stack Overflow as the primary site)
    SITE = StackAPI('stackoverflow')

    # Search for relevant questions
    questions = SITE.fetch('search/advanced',
                         q=question,
                         sort='votes',
                         accepted=True,
                         answers=1,
                         order='desc',
                         max_pages=1)

    # Process and format results
    results = []
    for q in questions['items'][:3]:  # Get top 3 results
        # Fetch the question with its answers
        full_q = SITE.fetch('questions/{0}'.format(q['question_id']),
                          filter='withbody',
                          site='stackoverflow')

        content = f"""Question: {q.get('title', '')}
URL: https://stackoverflow.com/questions/{q['question_id']}
Score: {q.get('score', 0)}
Content: {q.get('body', '')}"""

        results.append(Document(
            page_content=content,
            metadata={
                'source': 'stackoverflow',
                'url': f"https://stackoverflow.com/questions/{q['question_id']}",
                'score': q.get('score', 0)
            }
        ))

    return {"documents": results, "question": question}
```

ArXiv search function to fetch academic papers

```python
# ArXiv search function
def arxiv_search(state):
    """
    Search arXiv for research papers and publications.

    Args:
        state (dict): The current graph state

    Returns:
        state (dict): Updates documents key with arXiv results
    """
    print("---ARXIV SEARCH---")
    question = state["question"]

    # Configure arXiv client
    client = arxiv.Client(
        page_size=5,  # Number of results per query
        delay_seconds=3,  # Delay between API calls
        num_retries=3    # Number of retries on failed calls
    )

    # Create search query
    search = arxiv.Search(
        query=question,
        max_results=5,
        sort_by=arxiv.SortCriterion.Relevance
    )

    results = []

    try:
        # Fetch results
        for paper in client.results(search):
            # Extract and format the publication date
            if paper.published:
                pub_date = paper.published.strftime("%Y-%m-%d")
            else:
                pub_date = "Date not available"

            # Create content string with available information
            content_parts = [
                f"Title: {paper.title}",
                f"Authors: {', '.join(author.name for author in paper.authors)}",
                f"Published: {pub_date}",
                f"URL: {paper.entry_id}",  # arXiv URL
                f"PDF URL: {paper.pdf_url}",
                f"Abstract: {paper.summary[:500]}..."  # Truncate long abstracts
            ]

            # Add DOI if available
            if paper.doi:
                content_parts.insert(4, f"DOI: https://doi.org/{paper.doi}")

            content = '\n'.join(content_parts)

            # Create metadata
            metadata = {
                'source': 'arxiv',
                'title': paper.title,
                'url': paper.entry_id,
                'pdf_url': paper.pdf_url,
                'published_date': pub_date,
                'categories': [cat for cat in paper.categories],
                'doi': paper.doi if paper.doi else None
            }

            results.append(Document(
                page_content=content,
                metadata=metadata
            ))

    except Exception as e:
        print(f"Error during arXiv search: {str(e)}")
        # Add an error document if the search fails
        results.append(Document(
            page_content="Error retrieving arXiv results",
            metadata={'error': str(e)}
        ))

    return {"documents": results, "question": question}
```

# Define Workflow and Routing Logic

Set up routing logic to call the appropriate retrieval function

```python
def route_question(state):
    """
    Route question to appropriate search function.

    Args:
        state (dict): The current graph state

    Returns:
        str: Next node to call
    """
    print("---ROUTE QUESTION---")
    question = state["question"]
    source = question_router.invoke({"question": question})

    if source.datasource == "wiki_search":
        print("---ROUTE QUESTION TO WIKI SEARCH---")
        return "wiki_search"
    elif source.datasource == "vectorstore":
        print("---ROUTE QUESTION TO RAG---")
        return "vectorstore"
    elif source.datasource == "stack_exchange":
        print("---ROUTE QUESTION TO STACK EXCHANGE---")
        return "stack_exchange"
    elif source.datasource == "arxiv":
        print("---ROUTE QUESTION TO ARXIV---")
        return "arxiv"
```

```python
from langgraph.graph import END, StateGraph, START

workflow = StateGraph(GraphState)
# Define the nodes
workflow.add_node("wiki_search", wiki_search)  # web search
workflow.add_node("retrieve", retrieve)  # retrieve
workflow.add_node("stack_exchange", stack_exchange_search)
workflow.add_node("arxiv", arxiv_search)

# Build graph with conditional edges
workflow.add_conditional_edges(
    START,
    route_question,
    {
        "wiki_search": "wiki_search",
        "vectorstore": "retrieve",
        "stack_exchange": "stack_exchange",
        "arxiv": "arxiv",
    },
)

# Add edges to END
workflow.add_edge("retrieve", END)
workflow.add_edge("wiki_search", END)
workflow.add_edge("stack_exchange", END)
workflow.add_edge("arxiv", END)

# Compile
app = workflow.compile()
```

Testing for RAG based retrieval

```python
from pprint import pprint

def test_rag_retrieval(user_query):
    # Run
    print("\n=== Testing RAG Retrieval Route ===")
    inputs = {
        "question": user_query
    }

    # Stream the output
    for output in app.stream(inputs):
        for key, value in output.items():
            # Display each node
            pprint(f"Node '{key}':")
            # Optional: Uncomment to print full state at each node
            # pprint(value["keys"], indent=2, width=80, depth=None)
        pprint("\n---\n")

    # Display the final result
    pprint(value['documents'][0].dict()['metadata']['description'])

# Testing the modified function
test_rag_retrieval("What is an agent?")
```

Testing for WIKIPEDIA based retrieval

```python
from pprint import pprint

def test_wiki_retrieval(user_query):
    # Run
    print("\n=== Testing Wiki Retrieval Route ===")
    inputs = {
        "question": user_query
    }

    for output in app.stream(inputs):
        for key, value in output.items():
            # Node
            pprint(f"Node '{key}':")
            # Optional: print full state at each node
            # pprint.pprint(value["keys"], indent=2, width=80, depth=None)
        pprint("\n---\n")

    # Final generation
    pprint(value['documents'])

print("Testing Wiki Retrieval route...")
test_wiki_retrieval("Tell me about the Avengers")
```

```python
from stackapi import StackAPI

# Test Stack Exchange Route
def test_stack_exchange(user_query):
    print("\n=== Testing Stack Exchange Route ===")
    inputs = {
        "question": user_query
    }

    for output in app.stream(inputs):
        for key, value in output.items():
            pprint(f"Node '{key}':")
        pprint("\n---\n")

    # Print Stack Exchange results
    print("Stack Exchange Results:")
    for doc in value['documents']:
        result = doc.dict()
        pprint("Question Title: " + result['page_content'].split('\n')[0])
        pprint("URL: " + result['metadata']['url'])
        pprint("Score: " + str(result['metadata']['score']))
        print("\n")

# Run individual tests
print("Testing Stack Exchange route...")
test_stack_exchange("How to handle NullPointerException in Java?")
```

Testing for ARXIV based retrieval

```python
def test_arxiv(user_query):
    print("\n=== Testing arXiv Route ===")

    try:
        # Test with properly formatted input
        state = {
            "query": user_query
        }

        print("Testing Direct Search...")
        # Call the ArxivQueryRun instance
        response = arxiv.run(state)

        # Parse the response text
        if isinstance(response, str):
            # Split the response into sections
            sections = response.split('\n')

            # Extract and print publication info
            for line in sections:
                if line.startswith('Published:'):
                    print(f"\n{line}")
                elif line.startswith('Title:'):
                    print(line)
                elif line.startswith('Authors:'):
                    # Print complete author list
                    print(f"\nComplete Author List:")
                    authors = line.replace('Authors:', '').strip().split(', \n')
                    print(authors)
        else:
            print(response)

    except Exception as e:
        print(f"\nError during test execution: {str(e)}")
    finally:
        print("\n=== ArXiv Test Complete ===")

# Run test
print("\nTesting arXiv route...")
test_arxiv("Find recent research papers on transformer models")
```

# Combined Testing

```python
# Define the main function to prompt for user input, route the query, and call the appropriate test function

def process_user_query():
    # Step 1: Prompt the user for input
    query = input("Please enter your query: ")

    # Step 2: Initialize the state with the user's question
    state = {"question": query}

    # Step 3: Route the question using the `route_question` function
    route = route_question(state)

    # Step 4: Call the appropriate test function based on the identified route and print the result
    if route == "wiki_search":
        result = test_wiki_retrieval(query)
    elif route == "stack_exchange":
        result = test_stack_exchange(query)
    elif route == "arxiv":
        result = test_arxiv(query)
    elif route == "retrieve":
        result = test_rag_retrieval(query)
    else:
        result = "No suitable retrieval function found for the query."

    return result

process_user_query()
```
